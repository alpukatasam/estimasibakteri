```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Memanggil Dataset

```python
df = pd.read_csv('waterQuality1.csv')
```

# Visualisasi Data

```python
df=df.select_dtypes(include=[float,int])
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(),annot=True)
```

# Jumlah copper berdasarkan silver per liternya 

```python
models = df.groupby('silver').count()[['copper']].sort_values(by='copper',ascending=True).reset_index()
```

```python
fig = plt.figure(figsize=(15,5))
sns.barplot(x=models['silver'], y=models['copper'], color='royalblue')
plt.xticks(rotation=100)
```

# Jumlah bakteri berdasarkan virus.

```python
engine = df.groupby('viruses').count()[['bacteria']].sort_values(by='bacteria').reset_index()
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x=engine['viruses'], y=engine['bacteria'], color='royalblue')
```

# distribusi keamanan air perliternya

```python
df=df.select_dtypes(include=[float,int])
plt.figure(figsize=(15,5))
sns.distplot(df['uranium'])
```

# seleksi fitur

```python
features = ['aluminium','arsenic','barium','cadmium','chloramine','chromium','copper','flouride','viruses','lead','nitrates','nitrites','mercury','perchlorate','radium','selenium','silver','uranium']
##aluminium	arsenic	barium	cadmium	chloramine	chromium	copper	flouride	bacteria	viruses	lead	nitrates	nitrites	mercury	perchlorate	radium	selenium	silver	uranium
x = df[features]
y = df['bacteria']
x.shape, y.shape
```

# split data training dan data testing

```python
from sklearn.model_selection import train_test_split
x_train, X_test, y_train, y_test = train_test_split(x,y,random_state=70)
y_test.shape
```

# membuat model regresi linier

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
pred = lr.predict(X_test)
```

```python
score = lr.score(X_test, y_test)
print('akurasi model regresi linier = ', score)
```

# membuat inputan model regresi linier

```python
#'aluminium','arsenic','barium','cadmium','chloramine','chromium','copper','flouride','viruses','lead','nitrates','nitrites','mercury','perchlorate','radium','selenium','silver','uranium'
input_data = np.array([[1.65,0.04,2.85,0.007,0.35,0.83,0.17,0.05,0,0.054,16.08,1.13,0.007,37.75,6.78,0.08,0.34,0.02]])

prediction = lr.predict(input_data)
print('Prediksi Bakteri :', prediction)
```
